# DATA LOAD AND INITIAL EXPLORATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SAADAA_assessment_data - Sheet1.csv")
```

```python
print(df.info())
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], format="%Y-%m-%d", errors="coerce")  
```

# CUSTOMER SEGMENTATION ANALYSIS

```python
# Customer Value Segmentation

df["customer_value_segment"] = pd.qcut(df["order_value"], q=4, labels=["Low", "Mid-Low", "Mid-High", "High"])
```

```python
# Order Frequency Distribution

plt.figure(figsize=(8, 5))
sns.histplot(df["order_frequency"], bins=10, kde=True, color="blue")
plt.title("Distribution of Customer Order Frequency")
plt.xlabel("Orders per Customer")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

# Sales Performance Analysis

```python
# Monthly Sales Trends
df["order_month"] = df["order_date"].dt.month
monthly_sales = df.groupby("order_month")["order_value"].sum()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker="o", linestyle="-", color="b")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales (₹)")
plt.grid(True)
plt.show()
```

```python
# Top 5 Best-Selling Products
top_products = df["product_category"].value_counts().head(5)
print("Top 5 Best-Selling Products:")
print(top_products)
```

# Marketing Channel Effectiveness

```python
# Count of Orders from Each Marketing Source
marketing_performance = df["source"].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=marketing_performance.index, y=marketing_performance.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Effectiveness of Marketing Channels")
plt.xlabel("Marketing Channel")
plt.ylabel("Number of Orders")
plt.grid(axis="y")
plt.show()
```

# Return Analysis

```python
# Discounts vs. Return Rate
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["return_status"], y=df["discount_applied"], palette="coolwarm")
plt.title("Impact of Discounts on Returns")
plt.xlabel("Return Status")
plt.ylabel("Discount Applied (%)")
plt.grid(True)
plt.show()
```

```python
# Return Rate by Product Category
return_rate = df[df["return_status"] == "Yes"]["product_category"].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=return_rate.index, y=return_rate.values, palette="Reds_r")
plt.xticks(rotation=45)
plt.title("Return Rate by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Number of Returns")
plt.grid(axis="y")
plt.show()
```

# Inventory Risk Assessment

```python
# Count of Out-of-Stock Items
stockout_risk = df[df["inventory_status"] == "Out of Stock"]["product_category"].value_counts()
print("Stockout Risk (Out-of-Stock Items by Category):")
print(stockout_risk)
```

```python
# Inventory Shortage Prediction
low_stock = df[df["inventory_status"] == "Out of Stock"]
plt.figure(figsize=(10, 5))
sns.barplot(x=low_stock["product_category"].value_counts().index, 
            y=low_stock["product_category"].value_counts().values, palette="Oranges_r")
plt.xticks(rotation=45)
plt.title("Products at Risk of Stockouts")
plt.xlabel("Product Category")
plt.ylabel("Stockout Count")
plt.grid(axis="y")
plt.show()
```

This analysis provides actionable insights to enhance customer retention, optimize marketing spend, and improve inventory management. By implementing these recommendations, the company can increase revenue and improve operational efficiency.
